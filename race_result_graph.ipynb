### 2025 Belgian Grand Prix Race Results

**Import libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Import the CSV file**

```python
race_result = pd.read_csv('race_results.csv')
```

**Show the DataFrame**

```python
race_result
```

**Check the DataFrame Info**

```python
race_result.info()
```

**Change the datatypes from float to integer**

```python
race_result['Laps'] = race_result['Laps'].astype(int)
race_result['Points'] = race_result['Points'].astype(int)
```

**Change the data if the driver did not finished the race**

```python
race_result.loc[race_result['Laps'] != 44.0, 'Time'] = 'DNF'
```

**Check the result**

```python
race_result
```

**Check if the table has a null**

```python
race_result.isnull().sum()
```

**Describe the DataFrame**

```python
race_result.describe()
```

**Check if there's a duplicated data**

```python
race_result.duplicated().sum()
```

**Create a new column where shows the driver's abbrev**

```python
race_result.loc[:, 'Abbrev'] = (
    race_result['FullName']
    .fillna('')
    .astype(str)
    .apply(lambda s: s.split()[-1][:3].upper() if s else '')
)
```

```python
race_result
```

**Graph Shows the Points received by the driver after the race**

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x='Abbrev', y='Points', data=race_result)
ax.bar_label(ax.containers[0])
#plt.xticks(rotation=45, ha='right')
plt.title("Points Received by Each Driver After the Race")
plt.xlabel("Driver")
plt.ylabel("Points")
plt.tight_layout()
plt.show()
```

**Graph Shows the lap finished by the driver**

```python
finish_count = (race_result['Time'] != 'DNF').sum()
dnf_count = (race_result['Time'] == 'DNF').sum()

plt.figure(figsize=(6,6))
plt.pie([finish_count, dnf_count], labels=[f'Done 44 Laps ({finish_count})', f'43 Laps ({dnf_count})'])
plt.title('Lap Finished By Drivers')
plt.show()
```

**Graph Shows Driver's Time**

```python
plt.figure(figsize=(10,6))
plt.plot(race_result['Abbrev'], race_result['Time'], marker='o')
plt.title("Driver's Time")
plt.xlabel("Driver")
plt.ylabel("Laps")
plt.tight_layout()
plt.show()
```
